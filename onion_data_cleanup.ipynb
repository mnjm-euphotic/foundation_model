# Remove images with "crop" in their name and images with size 128x128

## Note target images will be "hard" linked

```python
from glob import glob
import os
from tqdm.auto import tqdm
import cv2
```

```python
dataset_path = "./dataset/onion"
target_dataset_path = "./dataset/onion_cleanup"
```

```python
src_imgs_l = glob(os.path.join(dataset_path, "*", "*", "*.jpg"))
assert len(src_imgs_l) > 0
len(src_imgs_l)
```

```python
for src_img_path in tqdm(src_imgs_l):
    img = cv2.imread(src_img_path)
    if "crop" in src_img_path or img.shape[:2] == (128, 128):
        continue
    tgt_img_path = os.path.join(target_dataset_path, os.path.sep.join(src_img_path.split(os.path.sep)[-3:]))
    dirpath = os.path.split(tgt_img_path)[0]

    print(f"{src_img_path} -> {tgt_img_path}")
    os.makedirs(dirpath, exist_ok=True)
    os.link(src_img_path, tgt_img_path)
```
